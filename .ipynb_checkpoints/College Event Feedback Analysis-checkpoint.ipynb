```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from textblob import TextBlob
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from wordcloud import WordCloud, STOPWORDS
from math import pi
from collections import Counter
import warnings
warnings.filterwarnings("ignore")
```

### STEP 2 â€” LOAD & EXPLORE DATA

```python
df = pd.read_csv("finalDataset0.2.csv")
```

```python
print("\n===== DATA OVERVIEW =====")
print(df.head())
print(df.info())
print(df.describe(include='all'))
```

### STEP 3 â€” DATA CLEANING

```python
df.drop_duplicates(inplace=True)
```

```python
text_cols = [c for c in df.columns if df[c].dtype == "object"]
df["feedback_text"] = df[text_cols].astype(str).agg(" ".join, axis=1)
```

```python
def clean_text(t):
    t = str(t).lower().strip()
    allowed = "".join(ch for ch in t if ch.isalnum() or ch.isspace())
    return allowed
```

```python
df["cleaned_feedback"] = df["feedback_text"].apply(clean_text)
df = df[df["cleaned_feedback"].str.strip() != ""]

print("ðŸ§¹ Cleaning complete!")
```

```python
df.head()
```

### STEP 4 â€” SENTIMENT ANALYSIS

```python
vader = SentimentIntensityAnalyzer()
```

```python
def blob_sentiment(t):
    pol = TextBlob(t).sentiment.polarity
    return pol
```

```python
def vader_sentiment(t):
    return vader.polarity_scores(t)["compound"]
```

```python
df["blob_score"] = df["cleaned_feedback"].apply(blob_sentiment)
df["vader_score"] = df["cleaned_feedback"].apply(vader_sentiment)
df["final_score"] = (df["blob_score"] + df["vader_score"]) / 2
```

```python
def label_sent(x):
    if x >= 0.05: return "Positive"
    elif x <= -0.05: return "Negative"
    return "Neutral"
```

```python
df["sentiment"] = df["final_score"].apply(label_sent)

print("Sentiment Analysis Done!")
print(df["sentiment"].value_counts())
```

### STEP 5 â€” STATISTICAL ANALYSIS

```python
rating_cols = [c for c in df.columns if df[c].dtype in [np.int64, np.float64]]

print("\nNumeric Rating Columns:", rating_cols)
```

```python
if rating_cols:
    print("\nRating Summary:")
    display(df[rating_cols].describe())
```

### STEP 6 â€” BEAUTIFUL VISUALIZATIONS

```python
# Donut Chart - (Sentiment Distribution)
sent_counts = df["sentiment"].value_counts()

colors = ["#7ED957", "#FF6F61", "#FFCC00"]

fig, ax = plt.subplots(figsize=(7, 7))
ax.pie(sent_counts, labels=sent_counts.index, autopct='%1.1f%%', colors=colors,
       wedgeprops={"linewidth": 2, "edgecolor": "white"})
centre = plt.Circle((0,0), 0.70, fc='white')
fig.gca().add_artist(centre)
plt.title("Sentiment Distribution", fontsize=16, weight="bold")
plt.show()
```

```python
# Voilin + Swarm Charts - (Sentiment vs Ratings)
for col in rating_cols[:4]:
    plt.figure(figsize=(8, 5))
    sns.violinplot(x="sentiment", y=col, data=df, inner=None, palette="pastel")
    sns.swarmplot(x="sentiment", y=col, data=df, color="black", alpha=0.6)
    plt.title(f"Distribution of {col.title()} by Sentiment", fontsize=15)
    plt.show()
```

```python
# Radar Chart - (Ratings)
if rating_cols:
    values = df[rating_cols].mean().tolist()
    values += values[:1]

    angles = [n / float(len(rating_cols)) * 2 * pi for n in range(len(rating_cols))]
    angles += angles[:1]

    fig = plt.figure(figsize=(8, 8))
    ax = plt.subplot(111, polar=True)

    ax.plot(angles, values, linewidth=2)
    ax.fill(angles, values, alpha=0.25)

    plt.xticks(angles[:-1], rating_cols, fontsize=12)
    plt.title("Average Ratings Comparison", size=18, weight="bold")
    plt.show()
```

```python
# Histogram - (Sentiment Intensity)
plt.figure(figsize=(8,5))
plt.hist(df["final_score"], bins=30, color="steelblue", alpha=0.8)
plt.axvline(df["final_score"].mean(), color="red", linestyle="--", label="Mean")
plt.title("Distribution of Sentiment Intensity", fontsize=16, weight="bold")
plt.xlabel("Final Sentiment Score")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
# Beautiful Word Clouds - (Positive)
positive_feedback = df.loc[df["sentiment"] == "Positive", "cleaned_feedback"]
positive_text = " ".join(positive_feedback.tolist())

positive_wc = WordCloud(
    width=900,
    height=450,
    background_color="white",
    colormap="Greens"
).generate(positive_text)

plt.figure(figsize=(10, 5))
plt.imshow(positive_wc, interpolation="bilinear")
plt.axis("off")
plt.title("Positive Feedback Word Cloud", fontsize=14, fontweight="bold")
plt.show()
```

```python
# Beautiful Word Clouds - (Negative)
negative_feedback = df.loc[df["sentiment"] == "Negative", "cleaned_feedback"]
negative_text = " ".join(negative_feedback.values)

negative_wc = WordCloud(
    width=900,
    height=450,
    background_color="white",
    colormap="Reds"
).generate(negative_text)

plt.figure(figsize=(10, 5))
plt.imshow(negative_wc, interpolation="bilinear")
plt.axis("off")
plt.title("Negative Feedback Word Cloud", fontsize=14, fontweight="bold")
plt.show()
```

### STEP 7 â€” INSIGHTS & RECOMMENDATIONS

```python
print("\n===== KEY INSIGHTS =====")

print(f"Positive: {sent_counts['Positive'] / len(df) * 100:.1f}%")
print(f"Neutral : {sent_counts['Neutral']  / len(df) * 100:.1f}%")
print(f"Negative: {sent_counts['Negative'] / len(df) * 100:.1f}%")

# Keyword extraction
def top_keywords(texts):
    words = " ".join(texts).split()
    return Counter(words).most_common(10)

print("\nTop Positive Keywords:", top_keywords(df[df["sentiment"]=="Positive"]["cleaned_feedback"]))
print("Top Negative Keywords:", top_keywords(df[df["sentiment"]=="Negative"]["cleaned_feedback"]))
```
